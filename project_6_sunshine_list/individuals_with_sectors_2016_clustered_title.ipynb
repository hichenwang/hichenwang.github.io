```python
import pandas as pd
import numpy as np
import statistics
%matplotlib inline
from altair import Chart
```

```python
df = pd.read_csv('individuals_with_sectors_2016_clustered.csv')
```

```python
df.head()
```

```python
df_cluster_job = df.groupby(['job_title', 'sector'])['job_title'].value_counts().groupby(level=1).nlargest(20).to_frame()
```

```python
df_cluster_job.rename(columns={"job_title": "count"}, inplace=True)
```

```python
df_cluster_job = df_cluster_job.reset_index(level=0)
```

```python
df_cluster_job = df_cluster_job.reset_index(level=0)
```

```python
df_cluster_job.rename(columns={"job_title": "job_title_cluster"}, inplace=True)
```

```python
df_cluster_job = df_cluster_job.reset_index(level=1)
```

```python
df_cluster_job.to_csv('job_titles.csv')
```
